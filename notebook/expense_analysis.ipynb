```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv('../data/expenses.csv', parse_dates=['Date'])
df.head()
```

```python
print("Total spent:", df['Amount'].sum())
print("\nSpending by Category:\n", df.groupby('Category')['Amount'].sum())
```

```python
plt.figure(figsize=(8,6))
df.groupby('Category')['Amount'].sum().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Spending Distribution by Category")
plt.ylabel("")
plt.show()
```

```python
df_daily = df.groupby('Date')['Amount'].sum()
plt.figure(figsize=(10,5))
df_daily.plot(marker='o')
plt.title("Daily Spending Trend")
plt.xlabel("Date")
plt.ylabel("Amount Spent")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()
pivot = df.pivot_table(index='Weekday', columns='Day', values='Amount', aggfunc='sum')

plt.figure(figsize=(12,6))
sns.heatmap(pivot, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap of Expenses by Day & Weekday")
plt.show()
```
